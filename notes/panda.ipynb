# Filter Data

```python
df3
```

```python
df3['last']=='Hitler'
# Marks the cells with the value matched as true
```

```python
#Let us apply this filer to an data frame
filt=(df3['last']=='Hitler')
df3[filt]
```

```python
df3.loc[filt,'email']
```

```python
# Filter using &(and) , | (or)

filt=((df3['last']=='Hitler') & (df3['first']=='Adolf'))
df3[filt]
```

```python
# To negate a filter
df3[~filt]
```

## Apply
Used for calling a function on our values
It works diffently on a dataframe and a series

```python
df3['EMAIL'].apply(len)
```

```python
def update_email(e):
    return e.upper()
df3['EMAIL'].apply(update_email)
```

```python
df3['EMAIL']=df3['EMAIL'].apply(update_email)
df3
```

```python
# when we use apply on a data frame
df3.apply(len)
```

Its giving us us the no. of rows in each columns
